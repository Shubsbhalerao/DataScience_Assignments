```python
mapping={'small':1,'large':2}
```

```python
df1 = df1.replace(mapping)
```

```python
X = np.array(df1.iloc[:,0:28])
y = np.array(df1.iloc[:,28])
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
X_norm = norm_func(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_norm,y,test_size=0.2,stratify=y,random_state=0)
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(12, input_dim=28, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train,validation_data=(x_test,y_test), epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(x_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions = model.predict_on_batch(x_train)
```

```python
for i in range(5):
    print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

```python
# evaluate the model
scores=model.evaluate(x_test,y_test)
scores
```

```python
history_df=pd.DataFrame(history.history)
history_df.head()
```

```python
# plot training and validation loss
plt.plot(history_df['loss'])
plt.plot(history_df['val_loss'])
plt.plot(history_df['val_accuracy'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train','Test'],loc='upper right');
```

```python
# plot the training and validation accuracy
plt.plot(history_df['accuracy'])
plt.plot(history_df['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train','Test'],loc='upper right');
```

# Hyper Parameter Tuning

```python
from sklearn.model_selection import GridSearchCV,KFold
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
```

```python
# write a user defined function to create a model
def create_model():
    model=Sequential()
    model.add(Dense(12,input_dim=8,activation='relu'))
    model.add(Dense(8,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    adam=Adam(lr=0.01)
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model
```

```python
model=KerasClassifier(build_fn=create_model,verbose=0)
```

```python
# define grid search parameters for batch size and epochs
batch_size=[10,20,40]
epochs=[10,50,100]
```

```python
# create a dictionary for grid search parameters
param_grid=dict(batch_size=batch_size, epochs=epochs)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = np.array(df1.iloc[:,0:28])
y = np.array(df1.iloc[:,28])
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
```

```python
# build and fit the grid search 
kfold=KFold(n_splits=3,shuffle=True,random_state=42)
grid=GridSearchCV(estimator=model,param_grid=param_grid,cv=kfold,verbose=5)
grid.fit(x_scaled,y)
```

```python
grid.best_params_
```

```python
# best score
grid.best_score_
```

```python
from tensorflow.keras.layers import Dropout
```

```python
def create_model(learn_rate,dropout_rate):
    model=Sequential()
    model.add(Dense(12,input_dim=8,activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(8,activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation='sigmoid'))
    adam=Adam(learning_rate=learn_rate)
    model.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
    return model
```
